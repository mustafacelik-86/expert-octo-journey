## kolon isimlerini düzenleyin, ilk 10 satırı iloc ile seçin

```python
df.columns = (
    df.columns
    .str.title()                                # İlk harfleri büyük yap
    .str.replace(r"\[1\]", "", regex=True)      # Nerede olursa olsun [1] ifadesini kaldır
    .str.replace("Un", "UN")                    # 'Un' → 'UN'
    .str.strip()                                # Baştaki ve sondaki boşlukları sil
)

df.columns
```

```python
#iloc ile ilk on satırı ve bütün sütunları aldık

df.iloc[:10,:]
```

## "Country or territory", "Population (1 July 2023)" sütunlarını loc ile seçin

```python
df.loc[:, ["Country Or Territory", "Population (1 July 2023)"]]
```

## 2023 nüfusu > 500 milyon üzerinde olan ülke bilgilerini yazdırın

```python
df[df["Population (1 July 2023)"] > 500_000_000]
```

##  Change (%) sütununu  + - % işaretlerinden kurtarın

```python
df["Change (%)"] = (
    df["Change (%)"]
    .astype(str)
    .str.replace("−", "-", regex=False)   # Unicode minus'u ASCII'ye çevir
    .str.replace("–", "-", regex=False)   # en-dash varsa onu da çevir
    .str.replace("%", "", regex=False)    # yüzde işaretini sil
    .str.replace("+", "", regex=False)    # artı işaretini sil
    .astype(float)
)
df["Change (%)"]
```

## Asya'da  + %1'in üzerinde artış olan ülke bilgilerini getiriniz

```python
df[(df["UN Continental Region"] == "Asia") & (df["Change (%)"] > 1)]
```

## Kıta gruplamasına göre en yüksek nüfuslu ülke ve en az nüfuslu ülke

```python
#en yüksek

df.loc[df.groupby("UN Continental Region")["Population (1 July 2023)"].idxmax(),
       ["UN Continental Region", "Country Or Territory", "Population (1 July 2023)"]]
```

```python
#en az

df.loc[df.groupby("UN Continental Region")["Population (1 July 2023)"].idxmin(),
       ["UN Continental Region", "Country Or Territory", "Population (1 July 2023)"]]
```

## Kıtalara göre %'sel ortalama nüfus değişimlerini pivot table ile hesapla

```python
df.pivot_table(index = "UN Continental Region", values = "Change (%)")

# aggfunc zaten mean olduğu için ayrıca yazmadım.
```

## Ülke isimlerinden köşeli parantez temizleme

```python
#bu regex metodunu bilmiyordum. Araştırarak öğrendim
df["Country Or Territory"] = df["Country Or Territory"].str.replace(r"\[.*?\]", "", regex=True).str.strip()
df.head(10)
```

## İçerisinde United geçen ülke isimlerini yazdırınız

```python
#uzamasın diye ilk beş adedini yazdırdım

df[df["Country Or Territory"].str.contains("United")].head()
```
