## 1b: Swarm Plot 

#### INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

## 2: Plot 12 most dangerous cities

```python
dan_cities = df.city.value_counts(ascending = False).head(12)
dan_cities
```

```python
values = dan_cities.values
values
```

```python
indexes = dan_cities.index
indexes
```

```python
plt.figure(figsize = (10,8))
ax = sns.barplot(x = indexes, y = values, palette = "husl")

for p in ax.containers:
    ax.bar_label(p, fmt="%d", padding=3)

ax.set_xticklabels(indexes, rotation=45, color="b")
plt.title("Most Dangerous Cities", color = "red")
plt.show()
```

```python
# istenen zm
```

```python
dan_cities
```

```python
indexes
```

```python
plt.figure(figsize = (10,8))
ax = sns.countplot(x = "city", data = df, order = indexes, palette = "husl")

for p in ax.containers:
    ax.bar_label(p, fmt="%d", padding=3)

ax.set_xticklabels(indexes, rotation=45, color="b")
plt.title("Most Dangerous Cities", color = "blue")
plt.show()
```

## 3: Plot 20 most dangerous states

```python
df.head()
```

```python
dan_states = df.state.value_counts(ascending = False).head(20)
dan_states
```

```python
values_states = dan_states.values
values_states
```

```python
indexes_states = dan_states.index
indexes_states
```

```python
plt.figure(figsize = (10,8))
ax = sns.barplot(x = indexes_states, y = values_states, palette = "husl")

for p in ax.containers:
    ax.bar_label(p, fmt="%d", padding=3)

ax.set_xticklabels(indexes_states, rotation=45, color="b")
plt.title("Most Dangerous States", color = "blue")
plt.show()
```

```python
# istenen kt
```

```python
plt.figure(figsize = (10,8))
ax = sns.countplot(x = "state", data = df, order = indexes_states, palette = "husl")

for p in ax.containers:
    ax.bar_label(p, fmt="%d", padding=3)

ax.set_xticklabels(indexes_states, rotation=45, color="b")
plt.title("Most Dangerous States", color = "blue")
plt.show()
```
