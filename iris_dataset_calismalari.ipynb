```python
#Taç yaprağı (petal length) en uzun 5 gözlemi çıkardık.

df.sort_values(by = "petal_length", ascending = False).head()
```

```python
#Çanak yaprağının (sepal width) en küçük ve en büyük boyutlarını tespit ettik.
df.agg({"sepal_width" : [max,min]})
```

```python
# Taç yaprağı 1.2 cm olan gözlemleri süzdük.

df[df.petal_length == 1.2]
```

```python
# Çanak yaprağı 3.9'dan büyük olanları süzdük.

df[df.sepal_width > 3.9]
```

```python
#setosa cinsi çiçeklerin verilerini çektik.

df[df.species == "setosa"]
```

```python
#tüm çiçek cinslerine göre sütunların min, max, mean değerlerini istedik.

df.groupby("species").agg(["max","min","mean"])
```

```python
#bu sefer sadece belli sütunların min, max, mean değerlerini istedik.

df.groupby("species")[["sepal_length", "petal_length"]].agg(["max","min","mean"])
```

```python
#species satırını baz alarak her sütun için farklı farklı isteklerde bulunduk.

df.groupby("species").agg({"sepal_length" : ["mean","max"], "sepal_width" : "mean", "petal_length" : "max", "petal_width" : "min"})
```

```python
#burada da benzer bir işlemi farklı bir yoldan yaptık.
df.groupby("species").mean(["sepal_length", "sepal_width"])
```

```python
#en büyük taç yaprak uzunluğu ile en kısa taç yaprak uzunluğu arasındaki farkı bulduk.

df.agg({"petal_length" : lambda x : x.max() -x.min()})
```

```python
# çiçek cinslerine göre sütunların max değerlerini istedik.

group = df.groupby("species")
group.max()
```

```python
#burada aslında iki üç sütun üzerinden de groupby yapmak isterdim ama dataset'in yyapısı çok müsaade etmiyor.
```
