```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl

import seaborn as sns 

import warnings;
warnings.filterwarnings("ignore") 
```

m değerini numpy kütüphanesi yardımı ile oluşturalım

```python
m = np.arange(11).astype(float)
```

```python
m
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
#istenen çıktı
```

```python
# Functional Method

plt.plot(m, E, "b", linewidth = 6)

plt.xlabel("M in g")
plt.ylabel("E in J")
plt.title("E=mc^2")

plt.show()
```

```python
# Object-oriented Method

fig, ax = plt.subplots()

ax.plot(m, E, "b", linewidth = 6)

ax.set_xlabel("M in g")
ax.set_ylabel("E in J")
ax.set_title("E=mc^2")

plt.show()
```

```python
#istenen çıktı
```

```python
#bunu sadece Functional Method ile çözüyorum:

plt.plot(m, E, "orange", linewidth = 4)

plt.xlabel("M in g")
plt.ylabel("E in J")
plt.title("E=mc^2")

plt.show()
```

## Alttaki görseli oluştururken farklı bir değişken tanımlamadan çözüyoruz.

```python
#istenen çıktı
```

```python
plt.plot(m, E, color = "green", linewidth = 4)

plt.xlabel("M in g")
plt.ylabel("E in J")
plt.title("E=mc^2")

plt.yscale("log")

plt.minorticks_on()
plt.grid(which="both", axis = "y", color = "gray", linestyle = "-", linewidth = 0.7, alpha = 0.7)

plt.show()
```

## Aşağıdaki değişkenlere göre ilgili plotlamaları yapalım.

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
#istenen çıktı
```

```python
plt.figure(figsize = (6, 4))
plt.plot(labels, july16_2007, color = "blue", label = "july16_2007")
plt.plot(labels, july16_2020, color = "orange", label = "july16_2020")

plt.legend(bbox_to_anchor = (1, 0.6), fontsize = 7)

plt.show()
```

```python
#istenen çıktı
```

```python
plt.figure(figsize = (6,4))
plt.plot(labels, july16_2007, color = "blue", label = "july16_2007")
plt.plot(labels, july16_2020, color = "orange", label = "july16_2020")

plt.legend(loc="center left", bbox_to_anchor=(1, 0.5), fontsize=7)

plt.show()
```

## Alttaki görselleri çizdirelim

```python
#İstenen görsel
```

```python
fig, ax1 = plt.subplots(figsize = (8,6))

ax1.plot(labels, july16_2007, color = "blue")
ax1.set_ylabel("July 16 2007", color = "blue", fontsize = 14)

ax2 = ax1.twinx()

ax2.plot(labels, july16_2020, color = "red")
ax2.set_ylabel("July 16 2020", color = "red", fontsize = 14)

plt.show()
```
