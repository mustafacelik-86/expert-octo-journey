```python
# 0'dan 9'a kadar sayılar oluştur (x ekseni verisi)
x = np.arange(0, 10)

# 2x2 (2 satır, 2 sütun) subplot oluştur, figür boyutu 12x6 inç
fig, ax = plt.subplots(2, 2, figsize=(12, 6))

# Tüm figüre başlık ekle
plt.suptitle("LineWidth & Marker & MarkerSize & Color")
# ------------------------------
# 1. Grafik (ax[0,0])
ax[0,0].plot(
    x,                 # x ekseni verisi
    "r",               # çizgi rengi: kırmızı ('r')
    marker="+",        # veri noktası işareti: artı simgesi
    markersize=5       # marker boyutu: 5
)
# ------------------------------
# 2. Grafik (ax[0,1])
ax[0,1].plot(
    x - 2,             # x verisinden 2 çıkar (grafik kayar)
    "y",               # çizgi rengi: sarı ('y')
    marker="o",        # veri noktası işareti: daire
    markersize=6       # marker boyutu: 6
)
# ------------------------------
# 3. Grafik (ax[1,0])
ax[1,0].plot(
    x - 3,             # x verisinden 3 çıkar
    "black",           # çizgi rengi: siyah (renk ismiyle)
    marker="s",        # veri noktası işareti: kare ('s')
    markersize=7,      # marker boyutu: 7
    lw=5               # çizgi kalınlığı (line width) = 5
)
# ------------------------------
# 4. Grafik (ax[1,1])
ax[1,1].plot(
    x - 4,             # x verisinden 4 çıkar
    "r",               # çizgi rengi: kırmızı ('r')
    marker="1",        # marker: aşağı doğru küçük üç çizgi
    ms=8               # marker size kısaltması: 8
)
# Alt grafikler arasındaki boşlukları ayarla
plt.tight_layout()
plt.show()
```

```python
# x verisi: -50 ile 50 arasında, 1'er artan tam sayılarla bir dizi oluşturuyor
x = np.arange(-50, 50)

# y1: x'in karesi (x^2) - Parabolik bir fonksiyon.
y1 = np.power(x, 2)

# y2: x'in küpü (x^3) - Kübik bir fonksiyon. Asimetrik ve üçgen benzeri bir şekil oluşturur.
y2 = np.power(x, 3)

# y3: x'in sinüsü (sin(x)) - Periodik bir fonksiyon, dalgalanma hareketi gösterir.
y3 = np.sin(x)

# y4: x'in kosinüsü (cos(x)) - Periodik bir fonksiyon, sinüs fonksiyonu ile benzer bir şekilde dalgalanır ancak faz farkı vardır.
y4 = np.cos(x)

# y5: x'in tanjantı (tan(x)) - Keskin dikey eğimlere sahip, periyodik bir fonksiyon. Aykırı değerler (outliers) olabilir.
y5 = np.tan(x)

# y6: x'in hiperbolik tanjantı (tanh(x)) - Hiperbolik bir fonksiyon, sinüs ve kosinüs fonksiyonlarına benzer şekilde dalgalanır ancak çıkışı -1 ile +1 arasında sınırlıdır.
y6 = np.tanh(x)

# y7: x'in hiperbolik sinüsü (sinh(x)) - Hiperbolik bir fonksiyon, sinüs fonksiyonuna benzer ancak çıkışı sürekli artan veya azalan bir şekilde büyük değerlere ulaşabilir.
y7 = np.sinh(x)

# y8: x'in hiperbolik kosinüsü (cosh(x)) - Hiperbolik bir fonksiyon, sürekli artan bir şekle sahiptir, ancak sinüs gibi dalgalanmaz.
y8 = np.cosh(x)

# y9: x'in üssü (exp(x)) - Üstel bir fonksiyon, sürekli artan bir şekil oluşturur ve büyüyen değerlere sahip olur.
y9 = np.exp(x)

fig1, ax1 = plt.subplots(nrows = 3, ncols = 3, figsize= (20, 20))
fig1.suptitle("Master Graphic", fontsize = 34)
ax1[0, 0].plot(x, y1, "tab:blue") # set the color of line chart with renk paleti
ax1[0, 0].set_title("Square Function") # setting title of subplot
ax1[0, 0].set_xlabel("X", fontsize = 18)
ax1[0, 0].set_ylabel("Y", fontsize = 18)

ax1[0, 1].plot(x, y2, "tab:orange") # set the color of line chart
ax1[0, 1].set_title("Cubic Function") # setting title of subplot
ax1[0, 1].set_xlabel("X", fontsize = 18)
ax1[0, 1].set_ylabel("Y", fontsize = 18)

ax1[0, 2].plot(x, y3, "tab:green") # set the color of line chart
ax1[0, 2].set_title("Sine Function") # setting title of subplot
ax1[0, 2].set_xlabel("X", fontsize = 18)
ax1[0, 2].set_ylabel("Y", fontsize = 18)

ax1[1, 0].plot(x, y4, "r-") # set the color of line chart
ax1[1, 0].set_title("Cosine Function") # setting title of subplot
ax1[1, 0].set_xlabel("X", fontsize = 18)
ax1[1, 0].set_ylabel("Y", fontsize = 18)

ax1[1, 1].plot(x, y5, "b-") # set the color of line chart
ax1[1, 1].set_title("Tangent Function") # setting title of subplot
ax1[1, 1].set_xlabel("X", fontsize = 18)
ax1[1, 1].set_ylabel("Y", fontsize = 18)

ax1[1, 2].plot(x, y6, "g-") # set the color of line chart
ax1[1, 2].set_title("Hyperbolic Tangent Function") # setting title of subplot
ax1[1, 2].set_xlabel("X", fontsize = 18)
ax1[1, 2].set_ylabel("Y", fontsize = 18)

ax1[2, 0].plot(x, y7, "m-") # set the color of line chart
ax1[2, 0].set_title("Hyperbolic Sine Function") # setting title of subplot
ax1[2, 0].set_xlabel("X", fontsize = 18)
ax1[2, 0].set_ylabel("Y", fontsize = 18)

ax1[2, 1].plot(x, y8, "y*") # set the color of line chart
ax1[2, 1].set_title("Hyperbolic Cosine Function") # setting title of subplot
ax1[2, 1].set_xlabel("X", fontsize = 18)
ax1[2, 1].set_ylabel("Y", fontsize = 18)

ax1[2, 2].plot(x, y9, "kv") # set the color of line chart
ax1[2, 2].set_title("Exp Function") # setting title of subplot
ax1[2, 2].set_xlabel("X", fontsize = 18)
ax1[2, 2].set_ylabel("Y", fontsize = 18)

plt.show()
```
