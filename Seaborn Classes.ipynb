## Scatter Plot

```python
sns.scatterplot(data = taxis, x = "fare", y = "tip")
plt.show()
```

## regression plot

```python
sns.regplot(data = taxis, x = "fare", y = "tip")
plt.show()
```

```python
# iki değişkenin arasındaki inceliyoruz.
# scatter'lara uygun olacak şekilde bir trend çizgisi oluşturuyor.
# regresyon: bağımsız değişken ve bağımlı değişken arasındaki ilişkiyi matematiksel olarak açıklama işlemi.
# daha önce aldığı verileri işleyerek, bütün datayı temsil edebilecek bir patern kuruyor. 
# gölgeler hata payını temsil eder. (ci = confident interval) default= %95
```

```python
plt.figure(figsize =(10,5))

sns.scatterplot(x = taxis["fare"], y = taxis["tip"])
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = taxis["fare"], y = taxis["tip"], hue = taxis["payment"])
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = taxis["fare"], y = taxis["tip"], hue = taxis["color"])
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = taxis["fare"], y = taxis["tip"], hue = taxis["passengers"])
plt.legend(loc = "lower right")
plt.show()
```

```python
# style, hue ile aynı işi yapar ama renklerle değil de şekillerle anlatır

plt.figure(figsize = (10,5))
sns.scatterplot(x = taxis["fare"], y = taxis["tip"], style = taxis["passengers"])
plt.legend(loc = "lower right")
plt.show()
```

```python
# hem style, hem hue

plt.figure(figsize = (10,5))
sns.scatterplot(x = taxis["fare"], y = taxis["tip"], style = taxis["passengers"], hue = taxis["color"])
plt.legend(loc = "lower right")
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize = (10,5))

sns.scatterplot(data = taxis, x ="fare", y = "tip", style = "color", ax = ax[0]) #birinci grafik olsun demek istedik
sns.scatterplot(data = taxis, x ="fare", y = "tip", hue = "passengers", ax = ax[1]) #ikinci grafik
plt.show()
```

```python
#renk paleti

plt.figure(figsize = (10,5))
sns.scatterplot(taxis, x = "fare", y = "tip", style = "passengers", hue = "color", palette = "viridis")
plt.legend(loc = "lower right")
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(taxis, x = "fare", y = "tip", style = "passengers", hue = "color", palette = "plasma")
plt.legend(loc = "lower right")
plt.show()
```

```python
#şeffafllık ayarı

plt.figure(figsize = (10,5))
sns.scatterplot(taxis, x = "fare", y = "tip", alpha = 0.5)
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(taxis, x = "fare", y = "tip", style = "color", hue = "passengers", palette = "tab10", alpha = 0.5, s = 100) #s = size
plt.show()
```
